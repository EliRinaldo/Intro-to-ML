```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/HamedTabkhi/Intro-to-ML/main/Dataset/D3.csv'

df = pd.read_csv(url)

print(df.head())
```

```python
X1 = df.values[:,0]
X2 = df.values[:,1]
X3 = df.values[:,2]
y = df.values[:,3]

LX1 = len(X1)
LX2 = len(X2)
LX3 = len(X3)
```

```python
X_0 = np.ones((LX1, 1))
X_1 = X1.reshape(LX1, 1)
X1 = np.hstack((X_0, X_1))

X_2 = X2.reshape(LX2, 1)
X2 = np.hstack((X_0, X_2))

X_3 = X3.reshape(LX3, 1)
X3 = np.hstack((X_0, X_3))

X1[:]
```

```python
theta = np.zeros(2)
theta
```

```python
def compute_cost(X, y, theta):
    predictions = X.dot(theta)
    errors = np.subtract(predictions, y)
    sqrErrors = np.square(errors)
    J = 1 / (2 * LX1) * np.sum(sqrErrors)
    return J

def gradient_descent(X, y, theta, alpha, iterations):
    Ly = len(y)  # Number of training examples
    cost_history = np.zeros(iterations)

    for i in range(iterations):
        predictions = X.dot(theta)
        errors = np.subtract(predictions, y)
        sum_delta = (alpha /  Ly) * X.transpose().dot(errors)
        theta -= sum_delta
        cost_history[i] = compute_cost(X, y, theta)

    return theta, cost_history
```

```python
#X1
iterations = 1500
alpha = .1
theta, cost_history = gradient_descent(X1, y, theta, alpha, iterations)
print('X1:')
print('Theta=')
print(theta)
print('Cost=')
print(cost_history)
print(f'Linear model: y = {theta[0]} + {theta[1]}*X1')

#X1 Data Plot
plt.scatter(X1[:,1], y, color='purple', label='Training Data')
plt.plot(X1[:,1], X1.dot(theta), color='red', label='Linear Regression')
plt.grid(True)
plt.xlabel('X1')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()

#X1 Loss Over Iteration Plot
plt.plot(range(1, iterations+1), cost_history, color='purple')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Convergence')
plt.show()
```

```python
#X2
theta, cost_history = gradient_descent(X2, y, theta, alpha, iterations)
print('X2:')
print('Theta=')
print(theta)
print('Cost=')
print(cost_history)
print(f'Linear model: y = {theta[0]} + {theta[1]}*X2')

#X2 Data Plot
plt.scatter(X2[:,1], y, color='lime', label='Training Data')
plt.plot(X2[:,1], X2.dot(theta), color='red', label='Linear Regression')
plt.grid(True)
plt.xlabel('X2')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()

#X2 Loss Over Iteration Plot
print(cost_history)
plt.plot(range(1, iterations+1), cost_history, color='lime')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Convergence')
plt.show()
```

```python
#X3
theta, cost_history = gradient_descent(X3, y, theta, alpha, iterations)
print('X3:')
print('Theta=')
print(theta)
print('Cost=')
print(cost_history)
print(f'Linear model: y = {theta[0]} + {theta[1]}*X3')

#X3 Data Plot
plt.scatter(X3[:,1], y, color='aqua', label='Training Data')
plt.plot(X3[:,1], X3.dot(theta), color='red', label='Linear Regression')
plt.grid(True)
plt.xlabel('X3')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.show()

#X3 Loss Over Iteration Plot
plt.plot(range(1, iterations+1), cost_history, color='aqua')
plt.rcParams["figure.figsize"] = (10, 6)
plt.grid(True)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Convergence')
plt.show()
```
